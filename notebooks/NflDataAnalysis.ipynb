# Funções

```python
def busca_jogador(nome_jogador: str, ano: str):
    df = pd.DataFrame()  

    caminho = f"../data/Players/{ano}"
    for arquivo in os.scandir(caminho):
        if arquivo.is_file() and arquivo.name.endswith(".json"):  
            try:
                tmp = pd.read_json(arquivo.path)  
                jogador = tmp[tmp["player"] == nome_jogador] 
                
                if not jogador.empty:  
                    df = pd.concat([df, jogador], ignore_index=True)

            except Exception as e:
                print(f"Erro ao ler {arquivo.name}: {e}") 

    return df 
```
